```python
import pandas as pd
import requests

def load_dataset(dataset_name, user='jarif87/DataSets'):
    url = f'https://raw.githubusercontent.com/{user}/main/{dataset_name}.csv'
    response = requests.get(url)
    response.encoding = 'utf-8'

    if response.status_code == 200:
        data = pd.read_csv(url)
        return data
    else:
        raise ValueError(f'Failed to load dataset {dataset_name} from {url}. Status code: {response.status_code}')

iris = load_dataset('C1')
print(iris.head())
```
